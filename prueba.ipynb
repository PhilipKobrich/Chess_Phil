```python
import copy 
chess=chess_class()
boards=[copy.deepcopy(chess.board)]
moves=[]
players=[]
for i in range(2):
    done,game_status=chess._check_game_end("white")
    if done:
        print(game_status)
        break

    while True:
        move=np.random.randint(4671)
        legal=chess._is_legal("white",move,chess.board)
        if legal:break
    moves.append(move)
    players.append("white")
    chess._take_move(move)
    boards.append(copy.deepcopy(chess.board))
    done,game_status=chess._check_game_end("black")
    if done:
        print(game_status)
        break

    while True:
        move=np.random.randint(4671)
        legal=chess._is_legal("black",move,chess.board)
        if legal:break
    moves.append(move)
    players.append("black")
    chess._take_move(move)
    boards.append(copy.deepcopy(chess.board))


```
